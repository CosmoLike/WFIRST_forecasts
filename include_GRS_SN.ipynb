```python
list=["GRS_nuisance","WF_3x2pt_clusterN_clusterWL_sys_opti","WF_SN"]


d1 = np.genfromtxt("/Users/timeifler/Dropbox/cosmolike_store/WFIRST_forecasts/chains/new/GRS_nuisance")
cov1=np.cov(d1[100000:,:],rowvar=False)


d2 = np.genfromtxt("/Users/timeifler/Dropbox/cosmolike_store/WFIRST_forecasts/chains/new/WF_3x2pt_clusterN_clusterWL_sys_opti")
cov2=np.cov(d2[100000:,:],rowvar=False) 


d3 = np.genfromtxt("/Users/timeifler/Dropbox/cosmolike_store/WFIRST_forecasts/chains/new/WF_SN")
cov3=np.cov(d3[100000:,:],rowvar=False)

```

```python
cov1_cosmo=cov1[0:7,0:7]
cov2_cosmo=cov2[0:7,0:7]
FM_SN=inv(cov3)

FMtot=np.zeros((7, 7))
FMtot[0,0]=FM_SN[0,0]
FMtot[3,3]=FM_SN[1,1]
FMtot[4,4]=FM_SN[2,2]
FMtot[6,6]=FM_SN[3,3]

FMtot[0,6]=FM_SN[0,3]
FMtot[6,0]=FM_SN[3,0]
FMtot[0,3]=FM_SN[0,1]
FMtot[3,0]=FM_SN[1,0]

FMtot[0,4]=FM_SN[0,2]
FMtot[4,0]=FM_SN[2,0]
FMtot[3,4]=FM_SN[1,2]
FMtot[4,3]=FM_SN[2,1]

FMtot[3,6]=FM_SN[1,3]
FMtot[6,3]=FM_SN[3,1]
FMtot[4,6]=FM_SN[2,3]
FMtot[6,4]=FM_SN[3,2]

# Add marginalized clusters and 3x2pt FMs, where marginzalized means only marginalized over systematics. Note that this is possible since the astrophysical 
# systematics are (assumed to be) independent. Projection into the w0wa space directly is not possible on the individual probe level
# See Albrecht et al 2006 p.96 for details.

FMtot+=inv(cov1_cosmo)+inv(cov2_cosmo)
cov_tot=inv(FMtot)

FM_noSN=inv(cov1_cosmo)+inv(cov2_cosmo)
cov_noSN=inv(FM_noSN)

##########full joint matrix ############
print "FM incl SN"
print FMtot
print "FM no SN"
print FM_noSN

covDE_noSN=cov_noSN[np.ix_([3,4],[3,4])]
FOMDE_noSN=np.power(det(inv(covDE_noSN)),0.5)
print("WFIRST DETF FoMDE_noSN=%e" %FOMDE_noSN)
print("sigma_w0=%e" %np.sqrt(covDE_noSN[0,0]))
print("sigma_wa=%e" %np.sqrt(covDE_noSN[1,1]))

covDE=cov_tot[np.ix_([3,4],[3,4])]
FOMDE=np.power(det(inv(covDE)),0.5)
print("WFIRST DETF FoM=%e" %FOMDE)
print("sigma_w0=%e" %np.sqrt(covDE[0,0]))
print("sigma_wa=%e" %np.sqrt(covDE[1,1]))
```
